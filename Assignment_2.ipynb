```python

# Remove units from attributes and rename columns
def remove_units(value):
    if isinstance(value, str):
        value = value.replace('kmpl', '').replace('CC', '').replace('bhp', '').replace('Lakh', '')
        return value
    return value

columns_to_convert = ['Mileage', 'Engine', 'Power']
for col in columns_to_convert:
    df[col] = df[col].apply(remove_units)


#Rename columns
df = df.rename(columns={'Mileage': 'Mileage_kmpl', 'Engine': 'Engine_CC', 'Power': 'Power_bhp'})

df.head()
```

```python

def split_car_name(name):
    parts = name.split(' ', 1)  # Split at the first space
    if len(parts) == 2:
        company = parts[0]
        model = parts[1]
        return pd.Series({'Company': company, 'Car_Model': model})
    else:
        return pd.Series({'Company': np.nan, 'Car_Model': name}) # Handle cases with no space


# Apply the function to the 'Name' column
df[['Company', 'Car_Model']] = df['Name'].apply(split_car_name)
df.head()
```

```python

import datetime

current_year = datetime.datetime.now().year

df['Car_Age'] = current_year - df['Year']


df.head()
```

```python

cols = list(df.columns)
cols.insert(cols.index('Name') + 1, cols.pop(cols.index('Company')))
cols.insert(cols.index('Name') + 2, cols.pop(cols.index('Car_Model')))
df = df[cols]
df.drop('Name', axis=1, inplace=True)

df.head()
```

```python

# Assuming you want to move 'Car_Age' after 'Year'

cols = list(df.columns)
cols.insert(cols.index('Year') + 1, cols.pop(cols.index('Car_Age')))
df = df[cols]

df.head()
```

```python

# Convert relevant columns to numeric, handling errors
for col in ['Mileage_kmpl', 'Engine_CC', 'Power_bhp', 'New_Price', 'Price']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python

# Replace null values in numerical columns with the mean
numerical_cols = ['Mileage_kmpl', 'Engine_CC', 'Power_bhp', 'Seats']
for col in numerical_cols:
    if col in df.columns:  # Check if the column exists
        mean_val = df[col].mean()
        df[col].fillna(mean_val, inplace=True)
```

```python
def convert_price(price_str):
    if isinstance(price_str, str):
        if 'Cr' in price_str:
            try:
                price_num = float(price_str.replace('Cr', ''))
                return price_num * 100  # 1 Cr = 100 Lakh
            except ValueError:
                return np.nan  # Handle cases where conversion fails
        elif 'Lakh' in price_str:
            try:
                return float(price_str.replace('Lakh', ''))
            except ValueError:
              return np.nan
        else:
            return price_str
    return price_str

# Change 'Price' to 'Price_Lakh' if 'Price' exists in the columns, otherwise use 'Price_Lakh'
# This ensures the code works whether 'Price' was renamed or not
col_to_convert = 'Price' if 'Price' in df.columns else 'Price_Lakh'
df[col_to_convert] = df[col_to_convert].apply(convert_price)

```

```python

# Convert 'New_Price' and 'Price' columns to numeric, handling errors
def convert_to_numeric(value):
    try:
        return float(str(value).replace(' Lakh', '').replace(',', ''))
    except (ValueError, AttributeError):
        return np.nan  # Return NaN for non-numeric values

if 'Price' in df.columns:  # Check if the 'Price' column exists
    df['Price'] = df['Price'].apply(convert_to_numeric)

if 'Price' in df.columns:
    df = df.rename(columns={'Price': 'Price_Lakh'})

df.head()
```

```python

# Remove "km/kg" from the 'Mileage' column
df['Mileage_kmpl'] = df['Mileage_kmpl'].astype(str).str.replace('km/kg', '', regex=False)
```

```python

# Convert 'Fuel_Type' and 'Transmission' to numerical representations
fuel_type_mapping = {'Diesel': 0, 'Petrol': 1, 'CNG': 2, 'LPG': 3}
transmission_mapping = {'Manual': 0, 'Automatic': 1}

df['Fuel_Type'] = df['Fuel_Type'].map(fuel_type_mapping)
df['Transmission'] = df['Transmission'].map(transmission_mapping)

df.head()
```
